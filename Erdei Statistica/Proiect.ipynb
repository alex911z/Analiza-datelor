```python
import pandas as pd
import numpy as np

def gini_index(values):
    sorted_values = np.sort(values)
    
    # Get the proportion of each value
    n = len(values)
    cumulative_sum = np.cumsum(sorted_values)  # Cumulative sum of sorted values
    total_sum = cumulative_sum[-1]  # Total sum of values
    
    # Calculate the Gini Index
    gini = (n + 1 - 2 * (cumulative_sum.sum() / total_sum)) / n
    return gini

# Load your dataset
df = pd.read_csv('heart.csv')  # Replace with your actual dataset path

# Loop through each column and calculate Gini Index
for column in df.columns:
    values = df[column].dropna()  # Remove any missing values
    gini_value = gini_index(values)
    print(f"Gini Index for '{column}': {gini_value:.4f}")
```

```python
import pandas as pd
import numpy as np
from collections import Counter

def entropy(values):
    # Count the frequency of each unique value
    value_counts = Counter(values)
    
    # Get the total number of values
    total_count = len(values)
    
    # Calculate probabilities (p(x_i))
    probabilities = [count / total_count for count in value_counts.values()]
    
    # Calculate the entropy using the formula
    entropy_value = -sum(p * np.log2(p) for p in probabilities if p > 0)
    return entropy_value

# Load your dataset
df = pd.read_csv('heart.csv')  # Replace with your actual dataset path

# Loop through each column and calculate entropy
for column in df.columns:
    values = df[column].dropna()  # Remove any missing values
    entropy_value = entropy(values)
    print(f"Entropy for '{column}': {entropy_value:.4f}")
```

```python
# Importing necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Load your dataset
df = pd.read_csv('heart.csv')  # Replace with your actual dataset path

# Prepare the feature matrix X and target vector y
X = df.drop('target', axis=1)  # Drop the target column to get the features
y = df['target']  # The target variable (0 or 1 for heart disease)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)

# Initialize the Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=50)  # n_estimators = 100 trees in the forest

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
predictions = model.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy:.4f}")

# Detailed performance report
print("\nClassification Report:")
print(classification_report(y_test, predictions))

# Optionally, display feature importance
print("\nFeature Importance:")
for feature, importance in zip(X.columns, model.feature_importances_):
    print(f"{feature}: {importance:.4f}")
```

```python
# Import necessary libraries
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report


df = pd.read_csv('heart.csv')  # Replace with your actual dataset path

# Prepare the feature matrix X and target vector y
X = df.drop('target', axis=1)  # Drop the target column to get the features
y = df['target']  # The target variable (0 or 1 for heart disease)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)

# Initialize the XGBoost model
model = xgb.XGBClassifier(
    n_estimators=100,           # Number of boosting rounds (trees)
    max_depth=3,                # Maximum depth of a tree
    learning_rate=0.1,          # Step size shrinking to prevent overfitting
    random_state=50
)

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
predictions = model.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy:.4f}")

# Detailed performance report
print("\nClassification Report:")
print(classification_report(y_test, predictions))

# Optionally, display feature importance
print("\nFeature Importance:")
xgb.plot_importance(model)
```
